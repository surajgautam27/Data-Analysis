```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sna
```

```python
batsman = pd.read_csv('../Matplotlib/content/sharma-kohli.csv')
batsman
```

```python
plt.plot(batsman['index'] , batsman['V Kohli'],color='green',linestyle='dotted',linewidth=2,marker='o', label='Virat')
plt.plot(batsman['index'] , batsman['RG Sharma'],color='black',linestyle='dashed',marker='D', markersize=2 ,label='Rohit')
plt.title('Rohit Sharma vs Virat Kolhi')
plt.xlabel('year')
plt.ylabel('Runs')
plt.legend(loc='best')
```
