```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.style.use('fivethirtyeight')
```

```python
data = sns.load_dataset('tips')
data
```

```python
sns.relplot(x='total_bill',y='tip' , data=data)
```

```python
sns.relplot(x='total_bill',y='tip',style='sex' ,hue='smoker',size='size',sizes=(15,200) ,data=data)
```

```python
sns.scatterplot(x='total_bill',y='tip',style='sex' ,hue='smoker',size='size' ,data=data)
```

```python
#For Catergorical Plots
#for categorical plot x data should be alphabetical
#y data should be numerical
```

```python
sns.catplot(x='day',y ='tip' , kind='strip',data=data) #jitter='jitter value'
```

```python
sns.catplot(x='day',y ='tip' , kind='swarm',hue='sex',data=data) #jitter='jitter value'
```

```python
#For Box Plot 
sns.boxplot(data['tip'])
```

```python
sns.boxplot(data['total_bill'])
```

```python
sns.catplot(x='day' ,y='tip' , kind='box',hue='sex', data=data)
```

```python
sns.catplot(x='day' ,y='total_bill', kind='violin' ,hue='sex',split=True,data=data)
```

```python
sns.catplot(x='smoker' , y='total_bill', kind='bar',data=data)
```

```python
sns.catplot(x='smoker' , y='total_bill',estimator=np.median,hue='sex' ,kind='bar',data=data)
```

```python
sns.catplot(x='sex' ,hue='smoker',kind='count',data=data)
```

```python
#heatmap
```

```python
flights=sns.load_dataset('flights')
flights
```

```python
flight_data=flights.pivot_table(index='year' ,columns='month', values='passengers',aggfunc='sum')
sns.heatmap(flight_data,cmap='summer',linewidth=0.5,cbar=False ,annot=True , fmt='d')
```

```python
sns.clustermap(flight_data,cmap='summer',metric='correlation',z_score=0,annot=True)
```

```python
#distplot
titanic=sns.load_dataset('titanic')
titanic
```

```python
titanic
```

```python
sns.distplot(titanic['age'],hist=False)
```

```python
sns.distplot(titanic['age'],kde=False,hist=False, rug=True)
```

```python
sns.distplot(titanic['age'],kde=False)
```
