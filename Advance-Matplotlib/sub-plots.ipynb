```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../Advance-Matplotlib/Dataset/batter.csv')
df
```

```python
plt.scatter(df['avg'],df['runs'])
plt.xlabel('Average')
plt.ylabel('Runs')
plt.title('Runs VS Average')
```

```python
fig, ax=plt.subplots( figsize=[13,9])
ax.scatter(df['avg'],df['runs'] ,marker='+')
ax.set_title('Average Vs Runs')
ax.set_xlabel('Average')
ax.set_ylabel('Runs')
```

```python
fig, ax=plt.subplots(nrows=2,ncols=1,figsize=[15,9], sharex=True)
ax[0].scatter(df['avg'],df['runs'] ,marker='+')
ax[0].set_title('Average Vs Runs')
ax[0].set_ylabel('Runs')
ax[1].scatter(df['avg'],df['strike_rate'] ,marker='+')
ax[1].set_xlabel('Average')
ax[1].set_ylabel('Strike Rate')
```

```python
fig, ax=plt.subplots(nrows=2,ncols=2,figsize=[15,9])
ax[0,0].scatter(df['avg'],df['runs'])
ax[0,1].scatter(df['avg'],df['strike_rate'] )
ax[1,0].hist(df['avg'])
ax[1,1].hist(df['runs'])
```
