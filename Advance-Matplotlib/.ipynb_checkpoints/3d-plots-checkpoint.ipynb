```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../Advance-Matplotlib/Dataset/batter.csv')
df
```

```python
fig=plt.figure()
ax=plt.figure(figsize=[15,12]).add_subplot(projection='3d')
ax.scatter(df['runs'],df['avg'],df['strike_rate'])
ax.set_title('Runs vs Avg vs Strike Rate')
ax.set_xlabel('Runs')
ax.set_ylabel('Average')
ax.set_zlabel('Strike Rate')
```

```python
x=[0,5,10]
y=[0,10,20]
z=[0,20,40]
fig = plt.figure()
ax=plt.figure().add_subplot(projection='3d')
ax.scatter3D(x,y,z, s=[100,100,100], color='black')
ax.plot(x,y,z,color='red')
```

```python
x=[0,5,10]
y=[0,0,20]
z=[0,20,40]
fig = plt.figure()
ax=plt.figure().add_subplot(projection='3d')
ax.scatter3D(x,y,z, s=[100,100,100], color='black')
ax.plot(x,y,z,color='red')
```

```python
#3d Surface Plot
x=np.linspace(-10,10,100)
y= np.linspace(-10,10,100)
xx,yy = np.meshgrid(x,y)
z=xx**2+yy**2
fig= plt.figure()
ax=plt.figure().add_subplot(projection='3d')
data=ax.plot_surface(xx,yy,z,cmap='viridis')
fig.colorbar(data)
```

```python
x=np.linspace(-10,10,100)
y= np.linspace(-10,10,100)
xx,yy = np.meshgrid(x,y)
z=xx**2+yy**2
fig= plt.figure()
ax=plt.figure().add_subplot()
data=ax.contour(xx,yy,z,cmap='viridis')
fig.colorbar(data)
```

```python
x=np.linspace(-10,10,100)
y= np.linspace(-10,10,100)
xx,yy = np.meshgrid(x,y)
z=xx**2+yy**2
fig= plt.figure()
ax=plt.figure().add_subplot()
data=ax.contourf(xx,yy,z,cmap='viridis')
fig.colorbar(data)
```

```python
data=pd.read_csv('../Advance-Matplotlib/Dataset/IPL_Ball_by_Ball_2008_2022.csv')
data
```

```python
temp_data=data[(data['ballnumber']<=6) & (data['batsman_run']==6)]
new_data=temp_data.pivot_table(index='overs', columns='ballnumber',values='batsman_run', aggfunc='count')
new_data
plt.figure(figsize=(15,10))
plt.imshow(new_data)
plt.colorbar()
plt.xticks(ticks=[0,1,2,3,4,5,],labels=[1,2,3,4,5,6])
plt.yticks(ticks=np.arange(0,20),labels=np.arange(1,21))
plt.show()
```
