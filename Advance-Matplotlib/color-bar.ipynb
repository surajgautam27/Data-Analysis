```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
plt.scatter(df['SepalLengthCm'],df['PetalLengthCm'])
```

```python
plt.scatter(df['SepalLengthCm'],df['PetalLengthCm'])
df['Species']=df['Species'].replace({'Iris-setosa':0,'Iris-virginica':1,'Iris-versicolor':2})
df
```

```python
batsman=pd.read_csv('../Advance-Matplotlib/Dataset/batter.csv')
batsman
```

```python
batsman=batsman.sample(50)
batsman
```

```python
plt.figure(figsize=(20,12))
plt.scatter(batsman['avg'],batsman['strike_rate'], s=batsman['runs'])
for i in range(batsman.shape[0]):
    plt.text(batsman['avg'].values[i],batsman['strike_rate'].values[i],batsman['batter'].values[i])
plt.show()
```

```python
plt.figure(figsize=(20,12))
plt.scatter(batsman['avg'],batsman['strike_rate'], s=batsman['runs'])
plt.axhline(130 ,color='blue')
plt.axvline(30,color='Blue')
plt.axhline(140,color='red')
for i in range(batsman.shape[0]):
    plt.text(batsman['avg'].values[i],batsman['strike_rate'].values[i],batsman['batter'].values[i])
plt.show()
```
