```python
data=df[['batsman','bowling_team','batsman_runs']]
mask=data['batsman']=='V Kohli'
data[mask].groupby('bowling_team')['batsman_runs'].sum().sort_values(ascending=False)
```

```python
def most_runs(batsman_name):
    mask=data['batsman']==batsman_name
    return data[mask].groupby('bowling_team')['batsman_runs'].sum().sort_values(ascending=False).head(3)
most_runs('S Dhawan')
```

```python
mask= df['over']>15
data=df[mask]
data=data.groupby('batsman').count()
data
played_balls=data['ball']>200
player=data[played_balls]
death_over_specialist=player['batsman_runs'].sort_values(ascending=False)
death_over_specialist.head(3)
```

```python
df
mask=df['over']>15
df=df[mask]
data=df.groupby('batsman')['batsman_runs'].count()
data=data[data>200]
player_list=data.index.tolist()
new_df=df[df['batsman'].isin(player_list)]
balls=new_df.groupby('batsman')['batsman_runs'].count()
runs=new_df.groupby('batsman')['batsman_runs'].sum()
strike_rate=(runs/balls)*100
strike_rate.sort_values(ascending=False)
```
